```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
product = LinearRegression()
```

```python
product.fit(X_train,y_train)
```

```python
product.score(X_train,y_train)
```

```python
y_pred = product.predict(X_test)
```

```python
y_pred
```
